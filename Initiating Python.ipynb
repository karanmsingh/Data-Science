```python
#user defined functions
def generate_data(n):
    Ïµ_values = []
    for i in range(n):
        e = np.random.randn()
        Ïµ_values.append(e)
    return Ïµ_values

data = generate_data(100)
plt.plot(data)
plt.show()
```

```python
#Using conditions
def generate_data(n, generator_type):
    Ïµ_values = []
    for i in range(n):
        if generator_type == 'U':
            e = np.random.uniform(0, 1)
        else:
            e = np.random.randn()
        Ïµ_values.append(e)
    return Ïµ_values

data = generate_data(100, 'U')
plt.plot(data)
plt.show()
```
