```python
import numpy as np
import pandas as pd 

%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_can = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
                       sheet_name='Canada by Citizenship', skiprows=20, skipfooter=2)
```

```python
df_can
```

```python
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'],axis=1,inplace=True)
df_can.head()
```

```python
df_can.rename(columns={'OdName':'Country','AreaName':'Continent','RegName':'Region'},inplace=True)
```

```python
df_can.head()
df_can.set_index('Country', inplace=True)
```

```python
df_can.columns=list(map(str,df_can.columns))
years=list(map(str,range(1980,2014)))
```

```python
df_japan = df_can.loc[['Japan'], years].transpose()
df_japan.head()
```

```python
df_japan.plot(kind='box', figsize=(8, 6))

plt.title('Box Plot of Japanese Immigration to Canada (1980-2013)')
plt.ylabel('Number of Immigrants')
plt.show()
```

```python
df_japan.describe()
```

```python
#Dataset for China and India
df_InChina = df_can.loc[['India','China'],years].transpose()
```

```python
df_InChina.head()
```

```python
df_InChina.describe()
```

```python
df_InChina.plot(kind='box', figsize=(8, 6))

plt.title('Indian and Chinese Immigration into Canada (1980-2013)')
plt.ylabel('Number of Immigrants')
plt.show()
```

```python
#Horizontal box plot
df_InChina.plot(kind='box', figsize=(10,7), color='blue', vert=False)

plt.title('Indian and Chinese Immigration into Canada (1980-2013)')
plt.xlabel('Number of Immigrants')
plt.ylabel('Country')
plt.show()
```

```python
#Using SUBPLOTS

#Use ARTIST Layer instead of the scripting layer

fig=plt.figure()
ax0=fig.add_subplot(1,2,1)     #(1 Row, 2Columns, 1st Plot)
ax1=fig.add_subplot(1,2,2)     #(",", 2nd Plot)

#Subplot: 1
df_InChina.plot(kind='box', color='blue',vert=False, figsize=(20,6),  ax=ax0)
ax0.set_title('Box Plot of Indian and Chinese Immigration into Canada (1980-2013)')
ax0.set_xlabel('Number of Immigrants')
ax0.set_ylabel('Country')

#Subpplot: 2
df_InChina.plot(kind='line',color=['Red','Green'],figsize=(20,6),ax=ax1)
ax1.set_title('Line Plot of Indian and Chinese Immigration into Canada (1980-2013)')
ax1.set_xlabel('Number of Immigrants')
ax1.set_ylabel('Years')

plt.show()
```

```python
#Question: Create a box plot to visualize the distribution of the top 15 countries (based on total immigration) grouped by 
#the decades 1980s, 1990s, and 2000s.
df_can['Total'] = df_can.sum(axis=1)
df_top15 = df_can.sort_values(['Total'], ascending=False, axis=0).head(15)
df_top15
```

```python
y80s=list(map(str, range(1980,1990)))
y90s=list(map(str,range(1990,2000)))
y20s=list(map(str,range(2000,2010)))
```

```python
df_80s = df_top15.loc[:, y80s].sum(axis=1)
df_90s = df_top15.loc[:, y90s].sum(axis=1)
df_20s = df_top15.loc[:, y20s].sum(axis=1)
```

```python
#Merging the series into 1 new DataFrame
new_df = pd.DataFrame({'1980s': df_80s, '1990s': df_90s, '2000s':df_20s}) 
new_df.head()
```

```python
new_df()
```

```python
new_df.plot(kind='box', figsize=(10, 6))
plt.title('Immigration from top 15 countries for decades 80s, 90s and 2000s')
plt.xlabel('Decades')
plt.ylabel('Number of Immigrants')
plt.show()
```
